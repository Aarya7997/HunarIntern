```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the given link dataset
url = '/content/Iris.csv'
iris = pd.read_csv(url)

print(iris.head())
```

```python
# 1. Bar Chart
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=iris)
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

# 2. Line Graph
plt.figure(figsize=(10, 6))
for Species in iris['Species'].unique():
    subset = iris[iris['Species'] == Species]
    plt.plot(subset.index, subset['SepalLengthCm'], label=Species)
plt.title('Sepal Length of Each Iris Species')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length')
plt.legend(title='Species')
plt.show()

# 3. Pie Chart
species_counts = iris['Species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Each Iris Species')
plt.show()

# 4. Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=iris)
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()

# 5. Histogram
plt.figure(figsize=(8, 6))
sns.histplot(iris['SepalLengthCm'], kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. Heatmap
iris_numerical = iris.drop(columns=['Species'])
correlation_matrix = iris_numerical.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()

# 7. Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```
